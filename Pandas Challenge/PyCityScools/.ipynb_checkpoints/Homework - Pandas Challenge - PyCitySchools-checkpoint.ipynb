```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools_count = school_data['school_name'].count()

#Calculate the total number of students
total_students_count = student_data['Student ID'].count()

#Calculate the total budget
total_budget = school_data['budget'].sum()

#Calculate the average math score
average_math_score = student_data['math_score'].mean()

#Calculate the average reading score
average_reading_score = student_data['reading_score'].mean()

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = (average_math_score + average_reading_score)/2

#Calculate the percentage of students with a passing math score (70 or greater)
students_passing_math_count = len(student_data.loc[student_data['math_score'] >= 70 , :])
percent_students_passing_math = (students_passing_math_count/total_students_count)*100

#Calculate the percentage of students with a passing reading score (70 or greater)
students_passing_reading_count = len(student_data.loc[student_data['reading_score'] >= 70 , :])
percent_students_passing_reading = (students_passing_reading_count/total_students_count)*100
```

```python
# Create a dataframe to hold the above results
district_df = pd.DataFrame({"Total Schools":[total_schools_count],
                           "Total Students" : [total_students_count],
                           "Total Budget" : [total_budget],
                            "Average Math Score":[average_math_score],
                            "Average Reading Score":[average_reading_score],
                            "% Passing Math":[percent_students_passing_math],
                            "% Passing Reading":[percent_students_passing_reading],
                            "Overall Passing Rate":[overall_passing_rate],
                            })
# Cleaner formatting
district_df['Total Students'] = district_df['Total Students'].map("{:,}".format)
district_df['Total Budget']  = district_df['Total Budget'].astype(float)
district_df['Total Budget']  = district_df['Total Budget'].map("${:,.2f}".format)

# District summary
district_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Grouping DataFrame By School-Name
school_group_df = school_data_complete.groupby(["school_name"])

# Average math score per school
average_math_score_per_school = school_group_df["math_score"].mean()

# Average reading score per school
average_reading_score_per_school = school_group_df["reading_score"].mean()

# Total budget per school
total_budget_per_school= school_group_df['budget'].mean()

# Total students per school
total_students_per_school = school_group_df['student_name'].count()

# Budget per student each school
budget_per_student = total_budget_per_school / total_students_per_school

# School Type
school_type_per_school = school_group_df['type'].unique()
school_type_per_school = school_type_per_school.astype('str')
school_type_per_school = school_type_per_school.replace("['District']","District")
school_type_per_school = school_type_per_school.replace("['Charter']","Charter")

# Number of students having passing math & reading score per school
student_passed_math_df_grouped = student_data.loc[student_data['math_score'] >= 70 , :].groupby('school_name').count()['math_score']
student_passed_reading_df_grouped = student_data.loc[student_data['reading_score'] >= 70 , :].groupby('school_name').count()['reading_score']

# Percentage of students passed math
percent_math_passed = (student_passed_math_df_grouped / total_students_per_school) * 100

# Percentage of students passed reading
percent_reading_passed = (student_passed_reading_df_grouped / total_students_per_school) * 100
```

```python
# Summarizing the Student Data per school

school_summary_df = pd.DataFrame({
                          "School Type": school_type_per_school,
                          "Total Students":total_students_per_school,
                          "Total School Budget": total_budget_per_school,
                          "Per Student Budget": budget_per_student,
                          "Average Math Score": average_math_score_per_school,
                          "Average Reading Score" : average_reading_score_per_school,
                          "% Passing Math" : percent_math_passed,
                          "% Passing Reading" : percent_reading_passed,
                          "% Overall Passing Rate" : (percent_math_passed + percent_reading_passed) / 2
                          })

school_summary_df_new = school_summary_df.copy()

# Formatting Data
school_summary_df_formatted = school_summary_df.copy()
school_summary_df_formatted['Total School Budget'] = school_summary_df_formatted['Total School Budget'].map("${:,.2f}".format)
school_summary_df_formatted['Per Student Budget'] = school_summary_df_formatted['Per Student Budget'].map("${:.2f}".format)

# School summary
school_summary_df_formatted
```

## Top Performing Schools (By Passing Rate)

```python
#Sort and display the top five schools in overall passing rate
sorted_school_summary_top = school_summary_df_formatted.sort_values('% Overall Passing Rate',ascending=False)
sorted_school_summary_top.head(5)
```

## Bottom Performing Schools (By Passing Rate)

```python
#Sort and display the five worst-performing schools
sorted_school_summary_worst = school_summary_df_formatted.sort_values('% Overall Passing Rate')
sorted_school_summary_worst.head(5)
```

## Math Scores by Grade
* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Dataframe of 9th, 10th, 11th and 12th grades - Math

math_9th_df = school_data_complete.loc[school_data_complete['grade'] == '9th',['Student ID','grade','school_name','math_score']]
math_10th_df = school_data_complete.loc[school_data_complete['grade'] == '10th',['Student ID','grade','school_name','math_score']]
math_11th_df = school_data_complete.loc[school_data_complete['grade'] == '11th',['Student ID','grade','school_name','math_score']]
math_12th_df = school_data_complete.loc[school_data_complete['grade'] == '12th',['Student ID','grade','school_name','math_score']]

# Grouping dataframe of each grade

math_9th_group = math_9th_df.groupby('school_name')
math_10th_group = math_10th_df.groupby('school_name')
math_11th_group = math_11th_df.groupby('school_name')
math_12th_group = math_12th_df.groupby('school_name')

# Calculating average math score for each grade - per school

math_avg_9th = math_9th_group['math_score'].mean()
math_avg_10th = math_10th_group['math_score'].mean()
math_avg_11th = math_11th_group['math_score'].agg(np.mean) 
math_avg_12th = math_12th_group['math_score'].agg(np.mean)
math_avg_df = pd.DataFrame({'9th':math_avg_9th,'10th':math_avg_10th,'11th':math_avg_11th,'12th':math_avg_12th})
math_avg_df
```

## Reading Score by Grade 

```python
# Dataframe of 9th, 10th, 11th and 12th grades - Reading

reading_9th_df = school_data_complete.loc[school_data_complete['grade'] == '9th',['Student ID','grade','school_name','reading_score']]
reading_10th_df = school_data_complete.loc[school_data_complete['grade'] == '10th',['Student ID','grade','school_name','reading_score']]
reading_11th_df = school_data_complete.loc[school_data_complete['grade'] == '11th',['Student ID','grade','school_name','reading_score']]
reading_12th_df = school_data_complete.loc[school_data_complete['grade'] == '12th',['Student ID','grade','school_name','reading_score']]

# Grouping dataframe of each grade

reading_9th_group = reading_9th_df.groupby('school_name')
reading_10th_group = reading_10th_df.groupby('school_name')
reading_11th_group = reading_11th_df.groupby('school_name')
reading_12th_group = reading_12th_df.groupby('school_name')

# Calculating average reading score for each grade - per school

reading_avg_9th = reading_9th_group['reading_score'].mean()
reading_avg_10th = reading_10th_group['reading_score'].mean()
reading_avg_11th = reading_11th_group['reading_score'].agg(np.mean)  # other way of calculating mean
reading_avg_12th = reading_12th_group['reading_score'].agg(np.mean)
reading_avg_df = pd.DataFrame({'9th':reading_avg_9th,'10th':reading_avg_10th,'11th':reading_avg_11th,'12th':reading_avg_12th})
reading_avg_df
```

## Scores by School Spending
* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Creating copy of dataframe

school_permonance = school_summary_df_new.copy()
school_permonance_resetindex = school_permonance.reset_index()

# Creating bins and group labels

spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_permonance_resetindex["Spending Ranges (Per Student)"] = pd.cut(school_permonance_resetindex['Per Student Budget'], spending_bins, labels = group_names)
school_permonance_resetindex_grouped = school_permonance_resetindex.groupby("Spending Ranges (Per Student)")

# Perfomance of schools by school spending
school_permonance_by_spending = school_permonance_resetindex_grouped.mean()
school_permonance_by_spending_df = school_permonance_by_spending[['Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', '% Overall Passing Rate']]
school_permonance_by_spending_df
```

## Scores by School Size

```python
# Creating bins and group labels

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_permonance_resetindex["School Size"] = pd.cut(school_permonance_resetindex['Total Students'], size_bins, labels = group_names)
school_permonance_resetindex_grouped_size = school_permonance_resetindex.groupby('School Size')

# Perfomance of schools by school size
school_permonance_by_size = school_permonance_resetindex_grouped_size.mean()
school_permonance_by_size_df = school_permonance_by_size[['Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', '% Overall Passing Rate']]
school_permonance_by_size_df
```

## Scores by School Type

```python
# Grouping dataframe by school type
school_permonance_resetindex_grouped_type = school_permonance_resetindex.groupby('School Type')

# Perfomance of schools by school type
school_permonance_by_type = school_permonance_resetindex_grouped_type.mean()
school_permonance_by_type_df = school_permonance_by_type[['Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', '% Overall Passing Rate']]
school_permonance_by_type_df
```

## Observable trends based on the data

1. Charter schools perfomred very good than District schools (based on overall passing rate)

2. Schools performance was good when average spending per student was less than $615 (based on average Spending Per Student), 
    otherwise gets worse when avg spending increases per student.

3. School performances gets poor when schools size becomes larger than 2000 students (based on number of students per school)
